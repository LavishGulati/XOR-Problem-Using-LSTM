```python
import pandas as pd
import numpy as np
from sklearn import model_selection
import random
import csv
```

```python
train = []
print(train)
```

```python
for i in range(100001):
    s=''
    parity=0
    for j in range(50):
        x=random.randint(1,101)
        if x%2!=0:
            parity+=1
        s += str(x%2)
#     print(s)
    l=[s,parity%2]
    train.append(l) 
train
```

```python
with open("train.csv","w") as f:
    writer=csv.writer(f,delimiter=",")
    writer.writerows(train)
```

```python
df = pd.read_csv("train.csv", delimiter=",")
Help=df.as_matrix(columns=df.columns[0:])
Help = Help.transpose()
X=Help[0,:]
Y=Help[1,:]
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
print(X.shape,Y.shape)
print(X)


```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
```
